## Example of Fourier Transform

This notebook contains some simple examples of taking Fourier transforms of 1-d signals. 

```python
from math import pi, sin, cos  
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_signal(signal, time, title):
    fig, ax = plt.subplots(figsize=(10,8))
    ax.plot(time, signal);
    ax.set_xlabel('Time in decimal fractions of pi');
    ax.set_ylabel('Signal amplitude')
    ax.set_title(title);

freq = pi/8
time = np.arange(0, 6*pi + freq, freq)
signal_1 = np.sin(time)

plot_signal(signal_1, time, 'Pure sine wave')
```

```python
def plot_fft(fft, freq, title):    
    fig, ax = plt.subplots(figsize=(10,8))
    frequency_axis = np.fft.fftfreq(len(fft), freq) 
    ax.plot(frequency_axis, fft);
    ax.set_xlabel('Frequency in decimal fractions of pi');
    ax.set_ylabel('Spectral amplitude');
    ax.set_title(title)

real_fft = np.absolute(np.fft.fft(signal_1))

plot_fft(real_fft, freq, 'Spectrum of pure sin tone')
```

```python
signal_2 = np.cos(time)

plot_signal(signal_2, time, 'Pure cosine wave')
```

```python
real_fft = np.absolute(np.fft.fft(signal_2))

plot_fft(real_fft, freq, 'Spectrum of pure cosin tone')
```

```python
freq = pi/16
time_2 = np.arange(0, 6*pi + pi/8, pi/16)
signal_3 = [sin(x) + cos(4*x) for x in time_2]

plot_signal(signal_3, time_2, 'Signal with two frequency tones')
```

```python
real_fft = np.absolute(np.fft.fft(signal_3))

plot_fft(real_fft, freq, 'Spectrum of two fewquency signal')
```
